```python
import json

f = open('schedule2021.json')

data = json.load(f)

f.close()

data
```

```python
import requests
i = 1
for week in data['weeks']:
    for game in week['games']:
        if game['home']['alias'] == 'HOU' or game['away']['alias'] == 'HOU':
            id = str(game['id'])
            url = 'http://api.sportradar.us/nfl/official/trial/v7/en/games/' + id + '/pbp.json?api_key=7b8w6jvqa6nehphcvb93ggvr'
            file_name = 'game' + str(i) + '.json'
            i+=1
            r = requests.get(url)
            with open(file_name, 'w') as outfile:
                json.dump(r.json(), outfile)
            print(url)
            print()
```
